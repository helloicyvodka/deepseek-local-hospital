```python
import pandas as pd
```

```python
dt = pd.read_excel("接入deepseek的医院-医院版.xlsx")
```

```python
dt["医院等级三分类"]=dt["医院评级"]
dt["医院等级三分类"].loc[dt["医院等级三分类"]=="三甲"]= '三级'
dt["医院等级三分类"].loc[dt["医院等级三分类"]=="三乙"]= '三级'
dt["医院等级三分类"].loc[dt["医院等级三分类"]=="二甲"]= '二级'
dt2 =dt[dt["医院等级三分类"].isin(['三级','二级','一级','未定级'])]
```

```python
dt2["医院等级三分类"].value_counts().to_frame()
```

```python
dt2.to_csv("接入deepseek的医院-医院版.csv",index=False)
```
